```python
import pandas as pd
import numpy as np

def balance_dataframe(df_final, target_col):
    df_min = df_final[df_final[target_col] == df_final[target_col].value_counts().idxmin()]
    df_maj = df_final[df_final[target_col] == df_final[target_col].value_counts().idxmax()]
    
    n_to_generate = len(df_maj) - len(df_min)
    synthetic_rows = []

    for _ in range(n_to_generate):
        sample_pair = df_min.sample(2).reset_index(drop=True)
        new_row = sample_pair.iloc[0].copy()

        for col in df_final.columns:
            if col == target_col:
                continue
            if np.issubdtype(df_final[col].dtype, np.number):
                new_row[col] = (sample_pair.iloc[0][col] + sample_pair.iloc[1][col]) / 2
            else:
                new_row[col] = np.random.choice([sample_pair.iloc[0][col], sample_pair.iloc[1][col]])

        synthetic_rows.append(new_row)

    df_synthetic = pd.DataFrame(synthetic_rows)
    df_synthetic[target_col] = df_min[target_col].iloc[0]

    df_balanced = pd.concat([df_final, df_synthetic], ignore_index=True)
    return df_balanced


df_final = balance_dataframe(df_final, "Life_Satisfaction_Bin3")
```

```python
def balance_multiclass(df, target_col):
    class_counts = df[target_col].value_counts()
    max_count = class_counts.max()
    
    balanced_parts = []

    for cls in class_counts.index:
        df_cls = df[df[target_col] == cls]
        n_to_generate = max_count - len(df_cls)

        synthetic_rows = []
        for _ in range(n_to_generate):
            sample_pair = df_cls.sample(2).reset_index(drop=True)
            new_row = sample_pair.iloc[0].copy()

            for col in df.columns:
                if col == target_col:
                    continue
                if np.issubdtype(df[col].dtype, np.number):
                    new_row[col] = (sample_pair.iloc[0][col] + sample_pair.iloc[1][col]) / 2
                else:
                    new_row[col] = np.random.choice([sample_pair.iloc[0][col], sample_pair.iloc[1][col]])

            synthetic_rows.append(new_row)

        df_synthetic = pd.DataFrame(synthetic_rows)
        df_synthetic[target_col] = cls

        balanced_df_cls = pd.concat([df_cls, df_synthetic], ignore_index=True)
        balanced_parts.append(balanced_df_cls)

    return pd.concat(balanced_parts, ignore_index=True)
df_final = balance_multiclass(df_final, "Life_Satisfaction_Bin3")
```
